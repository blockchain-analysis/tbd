```python
def normalize(df, column):
    return (df[column]-min(df[column]))/(max(df[column])-min(df[column]))
def normalize_std(df, column):
    return (df[column]-(df[column].mean()))/(df[column].std())
```

```python
df['avg_tx_frequency/min']=df_averages['avg_tx_frequency/min']
df['avg_tx_volume/min']=df_averages['avg_tx_volume/min']
df['avg_time_between_sends']=pd.to_timedelta(df_averages['avg_time_diff'])
df['total_sent'] = df_averages['total_sent']
# df['volume_mode'] = df_averages['volume_mode']

# df['volume_mode']= normalize(df, "volume_mode")
df['avg_tx_frequency/min']= normalize(df, "avg_tx_frequency/min")
df['avg_tx_volume/min']= normalize(df, "avg_tx_volume/min")
df['avg_time_between_sends']= normalize(df, "avg_time_between_sends")
df['total_sent']= normalize(df, "total_sent")

counts_df = pd.DataFrame(index=df.index)

counts_df['count'] = [ast.literal_eval(x)['count'] for x in df_txs['delta t+30']]

counts_df['count'] = normalize(counts_df, "count")

# counts_df["is_sus"] = 0
# counts_df.loc[counts_df['count'] > counts_df['count'].std() + counts_df['count'].mean(), "is_sus"] = 1
# counts_df.loc[counts_df["is_sus"] ==1]

plt.hist(counts_df["count"], bins=100)
plt.title("Histogram of the number of transactions that had a drop in the 30 minute returns")
plt.show()
counts_df["flag"] = 0
counts_df.loc[counts_df["count"] > 0.5, "flag"] = 1
```
